# Data Visualisation Project based on Employment,Unemployment and GDP in 2022

```python
# numpy → For numerical operations.
# pandas → For handling datasets.
# matplotlib.pyplot → For plotting graphs.
# seaborn → For advanced plotting 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the CSV file into a DataFrame called df.
df = pd.read_csv(r"C:\Users\asus\Downloads\Employment_Unemployment_GDP_data 2.csv")
df
```

```python
# here we use drop_duplicates function to remove the duplicates in data
df.drop_duplicates(inplace = True)
```

```python
#df.info() shows data types...
df.info()
```

```python
#df.isnull() shows a DataFrame of True/False indicating missing values.
print("Missing values in DataFrame:\n", df.isnull())
```

```python
#Get top 10 rows with highest agriculture employment and set that column as index.
Agriculture=df.nlargest(10, 'Employment Sector: Agriculture').set_index('Employment Sector: Agriculture')
country = Agriculture.apply('Country Name')# extract countries name with highest agriculture employment
print(country)
Top_10 = df['Employment Sector: Agriculture'].nlargest(10) #Extract top 10 values in the Agriculture column.

# Here we are preparing x and y for plotting.
x = country
y = Top_10
plt.figure(figsize = (10,8))
plt.title("Top 10 countries with highest employment in Agriculture sector")

plt.xlabel('Employment Rate')
plt.ylabel('Countries')

plt.barh(x,y,color="green")
plt.show()
```

```python
#Get top 10 rows with highest Industry employment and set that column as index.
Industry=df.nlargest(10, 'Employment Sector: Industry').set_index('Employment Sector: Industry')
country = Industry.apply('Country Name')# extract countries name with highest Industry employment
print(country)
Top_10 = df['Employment Sector: Industry'].nlargest(10)

# Here again we are preparing x and y for plotting.
x = country
y = Top_10
plt.figure(figsize = (20,8))
plt.title("Top 10 countries with highest employment in Industry sector")

plt.xlabel('Employment Rate')
plt.ylabel('Countries')

plt.bar(x,y,color="red")
plt.show()
```

```python
Unemployment = df.nsmallest(15, 'Unemployment Rate').set_index('Unemployment Rate')  #extracting 15 countries with minimum unemployment rate
country = Unemployment['Country Name']   #extracting country names
print(country)
#Get only the minimum 15 values from 'Unemployment Rate' column.
Top_10 = df['Unemployment Rate'].nsmallest(15)

# Here again we are preparing x and y for plotting.
x = country
y = Top_10
plt.figure(figsize = (20,8))
plt.title("15 countries with minimum unemployment rate")
plt.pie(y,labels=x)
plt.show()
```

```python
GDP_data=df.nlargest(10, 'GDP (in USD)').set_index('GDP (in USD)')  #extracting top 10 countries with highest gdp in USD
country = GDP_data.apply('Country Name')   #extracting country names
print(country)
Top_10 = df['GDP (in USD)'].nlargest(10)

# Here again we are preparing x and y for plotting.
x = country
y = Top_10
plt.figure(figsize = (20,8))
plt.title("Top 10 countries with highest GDP in 2022")
plt.pie(y,labels=x)
plt.show()
```
